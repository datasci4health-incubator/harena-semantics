```python
MODEL = 'fine_tuned/NER/ACD_15epochs'

MODEL_PATH = '../../models/word_embeddings/'

MODEL_PATH += MODEL
```

```python
import os, json

config_name = 'trainer_state.json'


#       "loss": 0.1488, 0.0821, 0.0557, 0.0426, 0.0381, 0.0182, 0.0188, 0.0137, 0.0101, 0.0093, 0.0051, 0.0052, 0.0049, 0.0029

config_file = os.path.join(MODEL_PATH, config_name)
config_json = json.load(open(config_file))


loss_values, validation_loss_values = [], []

# print(for i in config_json['log_history'])
# print(config_json['log_history'][0])
# loss_values = [ step for config_json['log_history'] ]

epochs = config_json['log_history']
# print(epochs)

for e in epochs:
    loss_values.append(e['loss'])
print(loss_values)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

# Use plot styling from seaborn.
sns.set(style='darkgrid')

# Increase the plot size and font size.
sns.set(font_scale=1.5)
plt.rcParams["figure.figsize"] = (12,6)

# Plot the learning curve.
plt.plot(loss_values, 'b-o', label="training loss")
plt.plot(validation_loss_values, 'r-o', label="validation loss")

# Label the plot.
plt.title("Learning curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.savefig('learning-curves')
plt.show()
```
