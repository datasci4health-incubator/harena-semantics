```python
import pysolr, os
import pandas as pd

# SOLR_URL = 'http://' + os.environ['SOLR_HOST'] + ':8983/solr/mesh'
SOLR_URL = 'http://solr:8983/solr/pmc'

# Create a client instance. The timeout and authentication options are not required.
solr = pysolr.Solr(SOLR_URL, always_commit=False)

# Do a health check.
solr.ping()
```

```python
# search_query = "A 58-year-old African-American woman presents to the ER with episodic pressing/burning anterior chest pain that began two days earlier for the first time in her life. The pain started while she was walking, radiates to the back, and is accompanied by nausea, diaphoresis and mild dyspnea, but is not increased on inspiration. The latest episode of pain ended half an hour prior to her arrival. She is known to have hypertension and obesity. She denies smoking, diabetes, hypercholesterolemia, or a family history of heart disease. She currently takes no medications. Physical examination is normal. The EKG shows nonspecific changes."
search_query = "sepsis"

filter = "body:epidemiology type:review"
query = 'title:' + search_query
related_papers = solr.search(q=query, fq=filter)
```

```python
retrieved_papers = dict()
titles = []
for paper in related_papers:
    print(paper)
    if paper.get('title') is not None:
        titles.append({'title': ''.join(paper.get('title'))})
retrieved_papers.update({'papers': titles})
```
