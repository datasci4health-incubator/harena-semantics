```python
## Simple WordCloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 

text = 'Lorem ipsum dolor sit amet, consectetur adipiscing elit. Quisque feugiat dignissim ligula sed feugiat. Fusce congue tincidunt risus, et maximus elit egestas non. In quis massa dapibus, bibendum magna sed, pellentesque risus. Quisque a tincidunt urna, quis pulvinar felis. Praesent est lacus, malesuada nec quam at, tincidunt laoreet felis. Sed. '
def generate_wordcloud(text): # optionally add: stopwords=STOPWORDS and change the arg below
    wordcloud = WordCloud(width=800, height=400, background_color="white",
                          relative_scaling = 1.0,
                          stopwords = {'to', 'of'} # set or space-separated string
                          ).generate(text)
    
    fig = plt.figure(1, figsize=(8, 4))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.axis("off")
    ## Pick One:
    # plt.show()
    plt.savefig("WordCloud.png")

generate_wordcloud(text)
```
